**디렉토리에서 원하는 파일목록 리스트로 가져오기**

- 16개 보의 기상 데이터와 수질 데이터가 연도별로 수집되어 있음.

```python
import os

def file_arr(path):
    arr = []
    fileList = os.listdir(path)
    fileList.sort()
    
    for f in fileList:
        # 임시 파일 제외
        if f.find('~$') != -1:
            continue
        if f.find('zNex~$hare') != -1:
            continue
        if f.find('.xlsx') != -1:
            arr.append(path / f)
    return arr
```

**연도별 수질 데이터 불러오기**

- 16개 보 수질/기상 데이터를 객체를 값으로 갖는 사전으로 불러오는 함수

```python
from collections import defaultdict

def data_merge(dirName):
    ndriver_weir_dict = defaultdict(dict)
    fileList = file_arr(base_path / dirName)
    
    for f in fileList:
        if '20' in f.name and '년' in f.name:
            year = f.name[f.name.find('20'): f.name.find('20') + 4]
        else:
            year = f'20{f.name[-7:-5]}'

        print(f'===== {year} 시작 =====')
            
        # pd.ExcelFile() 함수를 이용하여 엑셀 파일 정보와 함께 시트별로 데이터 불러옴
        with pd.ExcelFile(f) as xls:
            for sheetName in xls.sheet_names:
                if '(' in sheetName:
                    idx = sheetName.index('(')
                    weir = sheetName[:idx]
                elif '·' in sheetName:
                    weir = sheetName.replace('·', '')
                else:
                    weir = sheetName

                if weir in weirs:
                    ndriver_weir_dict[weir][year] = pd.read_excel(xls, sheet_name = sheetName)
                    print(f'{weir}', end = ', ')
            print()

    return ndriver_weir_dict
```

**수질/기상 데이터 통합**

```python
def merge_weir_df(weir):
    df_concat = pd.DataFrame()
    
    for year in years:
        df1 = jsb_wq[weir][str(year)]
        df2 = jsb_weather[weir][str(year)]
#         df3 = jsb_dflow[weir][str(year)]
        
        df1 = df1.astype({'측정일(시)': 'datetime64[ns]'})
        df2 = df2.astype({'일시': 'datetime64[ns]'}) 
#         df3 = df3.astype({'측정일(시)': 'datetime64[ns]'})
                                  
        df = pd.merge(df1, df2, left_on = '측정일(시)', right_on = '일시', how = 'left').drop(columns = '일시')
#         df = pd.merge(df, df3, how = 'left')
                                  
        df_concat = pd.concat([df_concat, df]).reset_index(drop = True)

    df_concat = df_concat.set_index('측정일(시)')
    return df_concat
```

세 특성의 경우에 결측치는 해당사항이 발생하지 않았음을 의미하는 것으로 이해할 수 있기 때문이다.

특성에 따라 결측치가 `ND` 또는 `-` 로 처리된 경우도 있다.
기본적으로 수치 데이터를 담아야 하는 특성임에도 불구하고 `object`를 `dtype`으로 갖는 특성을 대상으로
부동소수점이 아닌 데이터가 사용된 경우를 확인하면 다음과 같다

**죽산보 데이터**

```python
from collections import defaultdict

nan_values = defaultdict(set) # nan으로 사용된 문자열과 해당 문자열이 포함된 특성들의 집합으로 구성된 사전
object_columns = []

for col in jsb_data.columns:
    if jsb_data[col].dtype == 'object' and col != '지점명':  # object를 dtype으로 갖는 특성 대상. 지점명 제외.
        object_columns.append(col)
        print(col+":", end=" ") # 특성 이름 출력
        for value in jsb_data[col].value_counts().index: 
            try:
                float(value)
            except:
                nan_values[value].add(col)
                print(value, end=" ") # 수치형 데이터가 아닌 값만 출력
        print()
```

```python
object_columns
```

nan으로 사용된 문자열과 해당 문자열이 포함된 특성들은 다음과 같다.

```python
dict(nan_values)
```

nan 대신 사용된 값을 해당 특성에서 모두 `np.nan`으로 대체한 후에
해당 특성의 `dtype`을 `float64`로 지정한다.

```python
for nan_str in nan_values.keys():
    for col in nan_values[nan_str]:
        jsb_data.loc[:, col] = jsb_data[col].map(lambda x: np.nan if x == nan_str else x)
```

```python
jsb_data = jsb_data.astype({col:'float' for col in object_columns})
```

앞서 언급된 특성이 모두 부동소수점으로 구성되었음을 확인할 수 있다.

```python
jsb_data.info()
```

### 승촌보 데이터에 시프트 적용

**죽산보 데이터 기준으로 승촌보 데이터에 시프트 적용**

- 지정된 크기의 시프트를 승촌보 데이터에 대해서만 적용
- 이후 죽산보의 날짜 인덱스와 공통 인덱스를 마스크로 선택 후 적용
- 죽산보를 기준으로 승촌보의 시프트된 데이터 사이의 상관관계 확인 가능

```python
def jsb_scb_shifted(shift_size=1, 
                    year=2016, 
                    # month = 6,
                   ):

    # 승촌보 날짜 인덱스 시프트
    nShift_scb_daily = scb_df_daily.shift(periods=shift_size, freq='D')

    # 죽산보 데이터와 시프트된 승촌보 데이터의 공통 인덱스 활용
    # 죽산보 데이터 기준 승촌보의 시프트 이전 데이터를 동일 날짜로 지정
    jsb_index = jsb_df_daily.index
    scb_index = nShift_scb_daily.index
    common_index = jsb_index.intersection(scb_index)

    jsb_common = jsb_df_daily.loc[common_index]
    scb_common = nShift_scb_daily.loc[common_index]

    # 동일한 datetime 인덱스 사용 여부 확인
    assert (jsb_common.index == scb_common.index).all()

    # 죽산보 데이터와 shift된 승촌보 데이터의 연도별 상관관계

    # 지정된 년, 월 데이터 추출
    mask = (scb_common.index.year == year) # & (scb_common.index.month == month)

    jsb_common_ym = jsb_common.loc[mask]
    scb_common_ym = scb_common.loc[mask]

    # 동일 인덱스 사용 확인
    assert (scb_common_ym.index == jsb_common_ym.index).all()

    return jsb_common_ym, scb_common_ym
```

### 특성별 상관관계

다음 특성들을 대상으로 죽산보와 시프트된 승촌보 데이터 사이의 상관관계를 알아본다.

```python
common_features = ['EC', 'TN', 'DO', 'TOC', 'CHLA', 'pH', 'TP', 'Inflow', 'Discharge']
```

```python
shift_corr_dict = dict()

for year in range(2013, 2024):

    year_shift_corr_dict = dict()
    for shift_size in range(0, 6):

        jsb_common_ym, scb_common_ym = jsb_scb_shifted(shift_size, year)

        jsb_common_ym_features = jsb_common_ym[common_features]
        scb_common_ym_features = scb_common_ym[common_features]

        jsb_scb_common_corrwith = jsb_common_ym_features.corrwith(scb_common_ym_features)
        jsb_scb_common_corrwith = pd.DataFrame(jsb_scb_common_corrwith, columns=['Pearson'])

        year_shift_corr_dict[shift_size] = jsb_scb_common_corrwith

    year_shift_corr_df = pd.concat(year_shift_corr_dict, names=['shift_size', 'features'])
    year_shift_corr_df = year_shift_corr_df.unstack()
    year_shift_corr_df.columns = year_shift_corr_df.columns.droplevel(0)
    
    shift_corr_dict[year] = year_shift_corr_df
```

**죽산보 특성 기준: 0일부터 5일까지 시프트된 승촌보 특성 사이의 일대일 상관관계**

- 예를 들어, 승촌보의 `EC`를 0일부터 5일까지 시프트 시킨 후 각각의 경우에 대해
    죽산보의 `EC`와의 상관관계 측정
- 동일한 작업을 언급된 모든 특성에 대해 실행
- 2013년 데이터의 경우
    - 죽산보의 `EC`, `TN` 등: 각각 승촌보의 2~4일전 `EC`, `TN`와의 상관관계가 보다 높음.
    - 죽산보의 `CHLA`: 승촌보의 0~1일 전 `CHLA`와 상관관계 높음.
    - `Inflow`, `Discharge`: 두 보의 상관관계가 1에 가까움.
        즉, 승촌보의 방류량에 따라 죽산보의 방류량이 결정되는 것으로 보임.

```python
fig, axes = plt.subplots(11, 1, figsize=(10, 66))

for year in range(2013, 2024):
    ax = axes[year-2013]
    (shift_corr_dict[year].T).plot.bar(ax=ax, rot=0, xlabel=year, width=0.8)
    ax.set_ylim(-0.2, 1.05)
```

### 죽산보의 `CHLA` 기준 시프트된 승촌보 데이터 특성과의 상관관계

**`join_jsb_scb_shifted()` 함수**

- 지정된 죽산보 특성들과 시프트된 승촌보 특성들로 구성된 데이터프레임 생성
- 특성은 기본적으로 `common_features` 특성들만 대상으로 함.

```python
def join_jsb_scb_shifted(shift_size=1, 
                         year=2016, 
                         # month = 6,
                         features_jsb=['CHLA'],
                         features_scb=common_features):

    jsb_common_ym, scb_common_ym = jsb_scb_shifted(shift_size, 
                                                   year,
                                                  # month,
                                                  )

    # 죽산보의 CHLA를 기준으로 두 데이터셋 합병
    jsb_scb_joined = jsb_common_ym[features_jsb].join(scb_common_ym[common_features], lsuffix="_jsb")
    
    return jsb_scb_joined
```

**연도별 데이터프레임 생성**

- 연도별 데이터프레임: 죽산보의 `CHLA`와 시프트된 승촌보의 특성들 사이의 상관관계 포함
- 즉, 죽산보의 `CHLA`와 `common_features`에 포함된 특성들을 대상으로 상관관계 계산.
- 단, 시프트는 승촌보의 데이터에 대해서만 적용.

```python
shift_dict = dict()

for year in range(2013, 2024):
    
    year_shift_dict = dict()
    for shift_size in range(6):
        year_shift_dict[shift_size] = join_jsb_scb_shifted(shift_size, year, ['CHLA']).corr()['CHLA_jsb'][1:]
        year_shift_df = pd.concat(year_shift_dict, names=['shift_size', 'features']).unstack()

    shift_dict[year] = year_shift_df
```

```python
fig, axes = plt.subplots(11, 1, figsize=(10, 66))

for year in range(2013, 2024):
    ax = axes[year-2013]
    (shift_dict[year].T).plot.bar(ax=ax, rot=0, xlabel=f"Shifted correlation with CHPL of JSB ({year})", width=0.8)
    ax.set_ylim(-0.85, 0.85)
```
